```python
Neste notebook será apresentada uma análise exploratória com os dados de 2008 a 2019 do Índice de liberdade humana.
```

```python
#importando bibliotecas que serão utilizadas

import pandas as pd
import seaborn as sd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#carregando dados sobre o índice de liberdade humana

dados = pd.read_csv("hfi_cc_2021.csv")
dados
```

```python
#Criando um histograma para verificar a distribuição dos dados do índice geral

plt.hist(dados.iloc[:,4])
plt.title("Índice de liberdade humana")
plt.ylabel("Frequência")
plt.xlabel("Índice geral")
```

```python
#Entretanto esse histograma considera dados de todos os anos para todos os países, vamos criar um somente com os dados de 2019

dados2019 = dados[dados.year == 2019]
plt.hist(dados2019.iloc[:,4])
plt.title("Indíce de liberdade humana")
plt.ylabel("Frequência")
plt.xlabel("Índice geral")
```

```python
#Criando um boxplot e calculando média e desvio padrão desses dados

sd.boxplot(x = dados2019.hf_score).set_title("Índice Geral")
media2019 = np.mean(dados2019.hf_score)
desviopadrao2019 = np.std(dados2019.hf_score)
print("Média: ", media2019)
print("Desvio padrão: ", desviopadrao2019)
```

```python
#Nesta análise usaremos predominantemente a variável"hf_score", que representa o índice geral.
#Entretanto podemos repetir os passos para outras variáveis com pequenas alterações no código. Segue abaixo exemplo:

sd.boxplot(x = dados2019.pf_rol_civil).set_title("Boxplot para outra váriavel")
```

```python
#Além de agrupar os dados por ano, podemos ver a evolução de um mesmo país ao longo do tempo.

dadosbrazil = dados[dados.countries == "Brazil"]
dadosbrazil
```

```python
#E também criar gráficos para o país escolhido.

dadosbrazil.plot(x = "year", y = "hf_score", )
```

```python
#Devido a escala no eixo y, a impressão que o gráfico passa é de que houve uma grande queda do índice nesse período, o que não é verdade.
#Vamos arrumar a escala de forma a ter uma melhor visualização da evolução do índice.
dadosbrazil.plot(x = "year", y = "hf_score",).axis(ymin = 0, ymax = 10)
```

```python
#Podemos também responder algumas perguntas sobre os dados apresentados.
#Quais são os 10 países que obtiveram o maior indice geral em 2019?

top10_2019 = dados2019.sort_values("hf_score", ascending = False).head(10)
top10_2019.plot.bar(x = "countries", y = "hf_score")
top10_2019
```

```python
#E os 10 países com o menor índice?

bottom10_2019 = dados2019.sort_values("hf_score", ascending = True).head(10)
bottom10_2019.plot.bar(x = "countries", y = "hf_score")
bottom10_2019
```

```python
#Podemos também verificar se existe relação entre algumas variáveis. 
#Como exemplo usaremos os índices de inflação e de homicídios. 
#Os índices variam de 0 a 10, importante lembrar que uma possível correlação não implica em causalidade.

sd.regplot(data = dados2019, x = dados2019.pf_ss_homicide, y = dados2019.ef_money_inflation, fit_reg = False)
```

```python
#Por fim, realizaremos alguns testes para verificar se o índice geral é normalmente distribuído.
#Criando novamente o histograma com os dados de 2019

plt.hist(dados2019.iloc[:,4])
plt.title("Indíce de liberdade humana")
plt.ylabel("Frequência")
plt.xlabel("Índice geral")

#Vemos que o gráfico apresenta uma distribuição bastante assimétrica, porém podemos rodar mais alguns testes")
```

```python
#Criando um gráfico de probabilidade normal (qq plot)
fig, ax = plt.subplots()
stats.probplot(dados2019.hf_score, fit=True, plot=ax)
plt.show()

#Vemos que muitos pontos do gráfico se distanciam da linha. Indicando que não se trata de uma distribuição normal.
```

```python
#Realizando o teste de Shapiro-Wilk
#Caso encontremos um valor de p menor que 0.05 temos evidências para rejeitar a hipótese de distribuição normal.
stats.shapiro(dados2019.hf_score)
```
